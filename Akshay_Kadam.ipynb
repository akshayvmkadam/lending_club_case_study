### The objective of the analysis is to see who is likely to default in case of either fully paid or charged off loans and not current loan.

```python
loan_data = loan_data[loan_data.loan_status != "Current"]
loan_data.loan_status.unique()
```

### Missing values check

```python
(loan_data.isna().sum()/len(loan_data.index))*100
```

####  Columns with missing values are "emp_length", "revol_util". 
 - Checking the above columns before excluding them. 

```python
loan_data.info()
```

```python
print("Mode : " + loan_data.emp_length.mode()[0])
loan_data.emp_length.value_counts()
```

### As per the output above, mode value has far higher frequency than that of the next most frequent value. 
- This means that we can safely assign the value of mode to the null values in the column.
- Also the missing values are in very low percentage. So imputung with mode value doesnt affect the analysis much.

```python
loan_data.emp_length.fillna(loan_data.emp_length.mode()[0], inplace = True)
loan_data.emp_length.isna().sum()
```

```python
loan_data.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
loan_data.revol_util.isna().sum()
```

## Data Categorization
- "revol_util" and "int_rate" has continous values. So, we would be standardizing them
- "emp_length" --> { (< 1 year) is assumed as 0 and 10+ years is assumed as 10 }
- Although the datatype of "term" is plausibly an integer, there are only two values in the whole column and could be a categorical variable. 

```python
loan_data.revol_util = pd.to_numeric(loan_data.revol_util.apply(lambda x : x.split('%')[0]))
```

```python
loan_data.int_rate = pd.to_numeric(loan_data.int_rate.apply(lambda x : x.split('%')[0]))
```

```python
loan_data.emp_length = pd.to_numeric(loan_data.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

# Inferential Observations

### Conclusively, here are the driving factors(or driving variables)  behind loan default : 

- House ownership
    - Applicants who stay in rent
    - who have mortgaged their house along with loan of 14-15k 
    - who are applying for loan for home improvement having income slab   around 60k -70k.

- Income slab
    - Applicants who are eligible for 21-24% of rate of interest coming under the income slab of around 70k-80k.
    - who have already sanctioned a loan in the range 30k - 35k and are now eligible for 15-17.5% of rate of interest. 

- Loan amount
    - Applicants who have been already sanctioned with a loan for small business and now are eligible for loan greater than 14k
    - Applicants whose loan is sanctioned but amount is above 16k

- Grade
    - Applicants who are categorised under 'F'grade and are likely to be sanctioned a loan in the range of 15k-20k
    - Applicants who are categorised under 'G'grade and are eligible for 20% of rate of interest.
    - Other impacting grade would be 'B'.
    
- Number of accounts
    - Applicants who have multiple open accounts.

- Purpose of loan could be factor when its debt_consolidation
- Applicants who invests in the range of 5k-10k could be a risky.
- Applicants who apply or are eligible for 36 months as a loan repayment tenure.
- Sanctioning loan during year end
